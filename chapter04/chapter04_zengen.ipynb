```python
import requests

# URL からファイルをダウンロード
url = "https://nlp100.github.io/data/neko.txt"
response = requests.get(url)

# ダウンロードしたファイルをローカルに保存
with open("neko.txt", "wb") as f:
    f.write(response.content)

# MeCab タグを作成
mecab = MeCab.Tagger()

# ローカルファイルを読み込み、別のファイルに書き込む
with open("neko.txt", "r", encoding='utf-8') as f, open("neko.txt.mecab", "w", encoding='utf-8') as f2:
    lines = f.readlines()
    for text in lines:
        result = mecab.parse(text)
        f2.write(result)
```

**30. 形態素解析結果の読み込み**\
形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
with open("neko.txt.mecab", "r", encoding='utf-8') as f:
  general_list = []
  neko_list = []
  lines = f.readlines()
  for text in lines:
    neko_dic = {}
    suf = text.split("\t")
    if suf[0] == "EOS\n":
      continue
    temp = suf[1].split(',')
    neko_dic["surface"] = suf[0]
    if len(temp) <= 7:
      neko_dic["base"] = suf[0]
    else:
      neko_dic["base"] = temp[7]
    neko_dic["pos"] = temp[0]
    neko_dic["pos1"] = temp[1]
    neko_list.append(neko_dic)
    if suf[0]=="。":
      general_list.append(neko_list)
      neko_list = []
general_list #今後の問題で使います
```

**31. 動詞**\
動詞の表層形をすべて抽出せよ．

```python
suf_list = []
for sentense in general_list:
    for text in sentense:
        if text["pos"] == "動詞":
            suf_list.append(text["surface"])
verb = set(suf_list)
verb
```

**32. 動詞の基本形**\
動詞の基本形をすべて抽出せよ．

```python
base_list = []
for sentense in general_list:
    for text in sentense:
        if text["pos"] == "動詞":
            base_list.append(text["base"])
base_verb = set(base_list)
base_verb
```

**33. 「AのB」**\
2つの名詞が「の」で連結されている名詞句を抽出せよ．

```python
no_list = []
for sentense in general_list:
    for i in range(len(sentense)):
        if sentense[i]["surface"] == "の" and sentense[i-1]["pos"]=="名詞" and sentense[i+1]["pos"]=="名詞":
            no_list.append(sentense[i - 1]["surface"] + "の" + sentense[i + 1]["surface"])
no_list
```

**34. 名詞の連接**\
名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．

```python
noun_list = []
for sentense in general_list:
    count = 0
    sent = ""
    for i in range(len(sentense)):
        if sentense[i]["pos"] == "名詞":
            count += 1
            sent += sentense[i]["surface"]
        else:
            if count >= 2:
              noun_list.append(sent)
            count = 0
            sent = ""
noun_set = set(noun_list)
noun_set
```

**35. 単語の出現頻度**\
文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ．

```python
import collections
word_list = []
for sentense in general_list:
    for i in range(len(sentense)):
        if sentense[i]["pos"] != "補助記号" and sentense[i]["pos"] != "助詞" and sentense[i]["pos"] != "助動詞":
            word_list.append(sentense[i]["surface"])
c = collections.Counter(word_list)
c35 = c.most_common()
c35
```
