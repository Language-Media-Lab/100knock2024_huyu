### 4章 形態素解析

**30. 形態素解析結果の読み込み**

```python
with open("neko.txt.mecab", "r", encoding="utf-8_sig") as f:
    morphs = []
    morph = []
    for line in f.readlines():
        if not line.strip():
            continue
        if line == "EOS\n":
            morphs.append(morph)
            morph = []
            continue
        surface, rest = line.split("\t")
        if not surface:
            continue
        morph_list = rest.split(",")
        morph.append({
            "surface": surface,
            "base": morph_list[6],
            "pos": morph_list[0],
            "pos1": morph_list[1]
        })

display(morphs)
```

**31. 動詞**

```python
verb_surfaces = set()
for line in morphs:
    for morph in line:
        if morph["pos"] == "動詞":
            verb_surfaces.add(morph["surface"])
display(verb_surfaces)
```

**32. 動詞の基本形**

```python
verb_bases = set()
for line in morphs:
    for morph in line:
        if morph["pos"] == "動詞":
            verb_bases.add(morph["base"])
display(verb_bases)
```

**33. 「AのB」**

```python
no_nouns = []
for line in morphs:
    for i, morph in enumerate(line):
        try:
            if morph["pos"] == "名詞" and line[i+1]["surface"] == "の" and line[i+2]["pos"] == "名詞":
                no_nouns.append(morph["surface"] + line[i+1]["surface"] + line[i+2]["surface"])
        except:
            continue
display(no_nouns)
```

**34. 名詞の連接**

```python
nouns = []
noun = []
for line in morphs:
    for morph in line:
        if morph["pos"] == "名詞":
            noun.append(morph["surface"])
        else:
            if len(noun) > 1:
                nouns.append("".join(noun))
            noun = []
display(nouns)
```

**35. 単語の出現頻度**

```python
from collections import Counter

words = []
for line in morphs:
    for morph in line:
        if morph["pos"] == "記号":
            continue
        words.append(morph["base"])

count_words = Counter(words).items()
count_words = sorted(count_words, key=lambda x: x[1], reverse=True)

display(count_words)       
```

**36. 頻度上位10語**

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

top10_words = dict(count_words[:10])

plt.bar(top10_words.keys(), top10_words.values())
plt.title("出現頻度が高い上位10語")
plt.xlabel("単語")
plt.ylabel("出現頻度")
plt.ylim(0, 10000)
plt.show()
```

**37. 「猫」と共起頻度の高い上位10語**

```python
words_with_cat = []
for line in morphs:
    if "猫" not in [morph["base"] for morph in line]:
        continue
    for morph in line:
        if morph["pos"] == "記号" or morph["base"] == "猫":
            continue
        words_with_cat.append(morph["base"])

count_words_with_cat = Counter(words_with_cat).items()
count_words_with_cat = sorted(count_words_with_cat, key=lambda x: x[1], reverse=True)

top10_words_with_cat = dict(count_words_with_cat[:10])

plt.bar(top10_words_with_cat.keys(), top10_words_with_cat.values())
plt.title("「猫」と共起頻度の高い上位10語")
plt.xlabel("単語")
plt.ylabel("出現頻度")
plt.show()
```

**38. ヒストグラム**

```python
freq = [count for count in dict(count_words).values()]
plt.hist(freq, bins=100)
plt.title("単語の出現頻度分布")
plt.xlabel("出現頻度")
plt.ylabel("種類数")
plt.show()
```

**39. Zipfの法則**

```python
plt.xscale("log")
plt.yscale("log")
plt.plot(range(1, len(freq)+1), freq)
plt.title("単語の出現頻度と順位の両対数グラフ")
plt.xlabel("出現頻度順位")
plt.ylabel("出現頻度")
plt.show()
```
